```python
import pandas as pd
import numpy as np
import glob
import os
import datetime
from IPython.display import display
```

```python
sheet_name = ['策略分析','交易明細','交易分析','週期性分析','設定']
strategy_dict = dict()
```

```python
import os
arr = os.listdir("指數類")[1:]
for file in arr:
    #print(file)
    arr_c = os.listdir("指數類\\" + file)
    for file_c in arr_c:
        #print(file_c)
        arr_r = os.listdir("指數類\\" + file + "\\" + file_c)
        for xls in arr_r:
            if xls[:2] != "~$":
                print(xls)
                sheets = [file, file_c]
                for i in range(len(sheet_name)):
                    sheets.append(pd.read_excel("指數類\\" + file + "\\" + file_c + "\\" + xls, index_col = 0, header = 2, sheet_name=sheet_name[i]).dropna(how = 'all'))
                strategy_dict[xls[:-4]] = sheets
                
        
    print()
```

```python
strategy_list = []
for i in strategy_dict.keys():
    strategy = []
    strategy.append(i)
    strategy.append(strategy_dict[i][0])
    strategy.append(strategy_dict[i][1])
    strategy.append(strategy_dict[i][6].loc["壓縮"].values[0])
    strategy.append(strategy_dict[i][4].loc["平均獲利/平均虧損  比率"].values[0])
    strategy.append(strategy_dict[i][4].loc["平倉交易的平均K棒數"].values[0])
    strategy.append(strategy_dict[i][4].loc["交易總次數"].values[0])
    strategy.append(strategy_dict[i][6].loc["開始日期"].values[0])
    strategy.append(strategy_dict[i][6].loc["結束日期"].values[0])
    
    strategy_list.append(strategy)
```

```python
strategy_df = pd.DataFrame(strategy_list)
strategy_df.columns = ["檔名", "商品", "類型", "K棒週期", "賺賠比", "持倉K棒", "進場次數", "開始日期", "結束日期"]
```
